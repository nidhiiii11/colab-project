```python
#operators in python 
#' &' is bitwise operator 

titanic[(titanic['sex']=='female')&(titanic['survived']==1)].groupby('pclass')['survived'].count()

# 1 = 91
# 2 = 70
# 3 = 72
#   = 233
```
